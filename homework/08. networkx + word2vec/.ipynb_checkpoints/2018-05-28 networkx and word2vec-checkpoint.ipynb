```python
import sys
import gensim, logging
import networkx as nx
import matplotlib.pyplot as plt 

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
w = gensim.models.KeyedVectors.load_word2vec_format('araneum_upos_skipgram_300_2_2018.vec')
```

```python
w.init_sims(replace=True)
```

```python
verbs = ['смотреть', 'моргать', 'глазеть', 'пялиться', 'таращиться',
         'жмуриться', 'плакать', 'слушать', 'слышать', 'глотать',
         'жевать', 'молчать', 'говорить', 'кричать', 'дышать',
         'обонять', 'нюхать', 'чувствовать', 'осязать', 'ощущать']
```

```python
words = [verb + "_VERB" for verb in verbs]
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(verbs)
```

```python
for i in range(len(words)-1):
    for j in range(i+1, len(words)):
        if w.similarity(words[i], words[j]) > 0.5:
            G.add_edge(verbs[i], verbs[j])
print('done')
```

```python
pos=nx.nx.shell_layout(G)

plt.figure(figsize=(15, 10))
nx.draw_networkx_nodes(G, pos, node_color='green', node_size=20) 
nx.draw_networkx_edges(G, pos, edge_color='orange')
nx.draw_networkx_labels(G, pos, font_size=15, font_family='Arial')
plt.axis('off') 
plt.show()
```

```python
deg = nx.degree_centrality(G)
center = [nodeid for nodeid in sorted(deg, key=deg.get, reverse=True)[:5]]
print('Топ-5 центральных слов: ', ', '.join(center))
```

```python
count = 0
connectivity_components = [i for i in nx.connected_component_subgraphs(G)]
for component in connectivity_components:
    if len(component.nodes()) > 1:
        count += 1
        print('Компонента: ', ', '.join(sorted(component.nodes(), key=deg.get, reverse=True)),'\nРадиус: ', nx.radius(component))
print('Всего компонент: ',count)
```

```python
print(nx.average_clustering(G))
print(nx.transitivity(G))
```
